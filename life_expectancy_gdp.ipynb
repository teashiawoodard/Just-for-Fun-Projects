```python
#This project uses data on the life expectancy and gdp of different countries for 15 years. We will be comparing the different countries and working to find any relationship between the life expectancy and GDP of the countries. \
# A few questions that we will work to answer are:
#What countries do we have data on?
#What are the countries average GDP?
#Over the years, what country has increased GDP the most?
#Over the years, has the life expectancy of countries increased or decreased?
#Is there a relationship between life expectancy and GDP of countries?

```

```python
#Import important programs to use.
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import csv
```

```python
#import the data and save it to be used.
life_expectancy_gdp = pd.read_csv('all_data.csv')
```

```python
#Print out the first few rows of the data set
print(life_expectancy_gdp.head(10))
```

```python
#Print all the names of the countries we have data on
countries = life_expectancy_gdp['Country'].unique()
print('We have recorded data for life expectancy and GDP for the following countries for 15 years ' + str(countries) + '.')
```

```python
#Create a function that computes the average GDP for a country, then prints it out. 
def average_gdp(country, dataframe):
    country_gdp = life_expectancy_gdp.GDP[life_expectancy_gdp['Country'] == country]
    country_gdp_mean = country_gdp.mean()
    print('The average GDP for ' + country + ' is $' + str(round(country_gdp_mean, 2)) + " from 2000 to 2015.")

average_gdp('Chile', life_expectancy_gdp)
average_gdp('China', life_expectancy_gdp)
average_gdp('Germany', life_expectancy_gdp)
average_gdp('Mexico', life_expectancy_gdp)
average_gdp('United States of America', life_expectancy_gdp)
average_gdp('Zimbabwe', life_expectancy_gdp)
```

```python
#Create a visualization of the data to show the trends of each countries GDP over the course of 15 years.
chile_gdp = life_expectancy_gdp.GDP[life_expectancy_gdp['Country'] == 'Chile']
china_gdp = life_expectancy_gdp.GDP[life_expectancy_gdp['Country'] == 'China']
germany_gdp = life_expectancy_gdp.GDP[life_expectancy_gdp['Country'] == 'Germany']
mexico_gdp = life_expectancy_gdp.GDP[life_expectancy_gdp['Country'] == 'Mexico']
unitedstates_gdp = life_expectancy_gdp.GDP[life_expectancy_gdp['Country'] == 'United States of America']
zimbabwe_gdp = life_expectancy_gdp.GDP[life_expectancy_gdp['Country'] == 'Zimbabwe']
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]

plt.plot(years, chile_gdp, color = 'red')
plt.plot(years, mexico_gdp, color = 'green')
plt.plot(years, germany_gdp, color = 'black')
plt.plot(years, china_gdp, color = '#ffcc00')
plt.plot(years, unitedstates_gdp, color = 'blue')
plt.plot(years, zimbabwe_gdp, color = '#cc9900')
plt.legend(['Chile', 'Mexico', 'Germany', 'China', 'USA', 'Zimbabwe'])
plt.title('GDP by Country')
plt.xlabel('Years')
plt.ylabel('GDP')

plt.show()
plt.close()
```

```python
#Create a visualization of the data to show the trends of each countries life expectancy over the course of 15 years.
chile_life = life_expectancy_gdp['Life expectancy at birth (years)'][life_expectancy_gdp['Country'] == 'Chile']
china_life = life_expectancy_gdp['Life expectancy at birth (years)'][life_expectancy_gdp['Country'] == 'China']
germany_life = life_expectancy_gdp['Life expectancy at birth (years)'][life_expectancy_gdp['Country'] == 'Germany']
mexico_life = life_expectancy_gdp['Life expectancy at birth (years)'][life_expectancy_gdp['Country'] == 'Mexico']
unitedstates_life = life_expectancy_gdp['Life expectancy at birth (years)'][life_expectancy_gdp['Country'] == 'United States of America']
zimbabwe_life = life_expectancy_gdp['Life expectancy at birth (years)'][life_expectancy_gdp['Country'] == 'Zimbabwe']
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015]
```

```python
#Create a visualtization that groups data by country and showcases the life expectancy.
sns.boxplot(x = 'Life expectancy at birth (years)', y = 'Country',  data = life_expectancy_gdp, hue = 'Country', palette = 'bright')
plt.title('Life Expectancy by Country')
plt.xlabel('Life Expectancy at Birth (in Years)')
plt.show()
plt.close()
```

```python
#Create a line graph that showcases the life expectancy of each country by year.
ax1 = plt.subplot(3, 2, 1)
plt.plot(years, chile_life, color = 'red')
plt.xlabel('Years')
plt.ylabel('Life Expectancy')
ax2 = plt.subplot(3, 2, 2)
plt.plot(years, mexico_life, color = 'green')
plt.xlabel('Years')
plt.ylabel('Life Expectancy')
ax3 = plt.subplot(3, 2, 3)
plt.plot(years, germany_life, color = 'black')
plt.xlabel('Years')
plt.ylabel('Life Expectancy')
ax4 = plt.subplot(3, 2, 4)
plt.plot(years, china_life, color = '#ffcc00')
plt.xlabel('Years')
plt.ylabel('Life Expectancy')
ax5 = plt.subplot(3, 2, 5)
plt.plot(years, unitedstates_life, color = 'blue')
plt.xlabel('Years')
plt.ylabel('Life Expectancy')
ax6 = plt.subplot(3, 2, 6)
plt.plot(years, zimbabwe_life, color = '#cc9900')
plt.legend(['Chile', 'Mexico', 'Germany', 'China', 'USA', 'Zimbabwe'])
plt.suptitle('Life Expectancy By Country')
plt.xlabel('Years')
plt.ylabel('Life Expectancy')
plt.subplots_adjust(hspace = 0.5, wspace = 0.5)
plt.show()
plt.close()
```

```python
#Create a visualization of the data to visualize if there is any correlation between GDP and life expectancy.
sns.scatterplot(x = 'GDP' , y = 'Life expectancy at birth (years)', data = life_expectancy_gdp, hue = 'Country', palette = 'bright')
plt.title('Life Expectancy Per Country Vs GDP')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy')
plt.show()
plt.close()
```
